```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

# --- Midprice Process Generation ---
S0 = 1.2
mu = 0.0001
sigma = 0.002

n_steps = 390
T = 1
dt = T / n_steps
N = 1
dW = np.random.standard_normal(size=(n_steps, N))

S = np.zeros(n_steps)
S[0] = S0
for t in range(1, n_steps):
    noise = dW[t, 0]          # use scalar noise
    S[t] = S[t-1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * noise)

start_date = '2025-10-27 09:30:00'
end_date   = '2025-10-27 15:59:00'
minutes_index = pd.date_range(start=start_date, end=end_date, freq='min')

S_series = pd.Series(S, index=minutes_index, name='PX')
S_df = pd.DataFrame({'Mid': S_series}, index=minutes_index)

# --- Bid-Ask Spread Simulation ---

S_df['spread'] = 0.0002

# --- Isolating Random Samples ---
random_sample_df = S_df.sample(n = 5)
random_sample_df
random_times = random_sample_df.index.tolist()
```
