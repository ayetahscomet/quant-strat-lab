### Date:       15-Jan-2026

### Date:       14-Jan-2026

**Global Economy:   Yen tumbles and Japan stocks soar as election talk reignites ‘Takaichi trade’**

*Leo Lewis in Tokyo, Ian Smith in London and William Sandlund in Hong Kong*

- The Japanese markets are being dominated by roumours that Prime Minister Sane Takaichi will dissolve parliment on 23-Jan-2026, to call a snap election for 08-Feb-2026; The move aims to leverage her high approval ratings and solidify her party's majority; Investors are interpreting as a green lighht for aggressive fiscal stimulus and a pivot away from rapid monetary tightening

### Date:       13-Jan-2026

**UK Retail:    UK retail spending rose at lowest pace for 7 months in December**

*Valentina Romei and Philip Stafford in London*

- UK retail spending continued to growth slowly in December, below the rate of inflation, showing that the christmas season still couldn't jumpstart consumer spending; The value of retail sales increased by an annual rate of 1.2 per cent, down from 1.4 per cent in November (a new low post May's 1 per cent)
- Consumer spending in the festive season was mainly spurred by food (a Christmas un-negotiables), with spending on gifting items such as electronics, furniture, and clothing, significantly underperformed
- Retail bosses have blamed weeks of speculation over Reeves' tax rises; The value of sales rose by 2.3 per cent across 2025, according to BRC data, which is below the latest inflation rate of 3.4 percent, showing that consumers generally cut spending quantities
- A sharp slowdown in inflation in the first half of 2026, along with further rate cuts by the BoW should provide consumers a little bit of stability, and unlock real spending power

### Date:       10-Jan-2026

**UK government spending:   Spending on consultants cut by 14% across UK government**

- UK Government cut consulting costs by 14 per cent in the year ending March 2025, showing a significant attempt to reduce government spending; Home Office, in particular, reduced annual spending by 54 per cent
- A lack of government skills investment encouraged the reliance on external advisors over recent years, with civil servant numbers hitting a 20-year high; tministers have also promised to create "a leaner, higher skilled" workforce, allocating GBP 50 mn in civil service transformation funds, cutting the use of external adisors, whilst not encouraging civil servant layoffs (which would hike the unemployment rate significantly); however, they do aim to cut 1 in 10 roles by 20230
- The Department for Energy Security and Net Zero recorded the largest real-terms rise in consultancy spending of £33mn, alongside a £11mn rise in temporary worker costs; The health department accounted for the majority of the decline, delivering a GBP 1.1 bn reduction in 2024-25
- Cuts to Home Office spending follow a 10-fold increase in consultancy costs during the last parliament, driven by efforts to tackle small boat crossings and implement the Rwanda plan; This government is relentlessly rooting out waste to protect taxpayer money and make the state more efficient; This should be reflected in long-dated government bond prices




### Date:       09-Jan-2026

**Technology sector:    Smartphone and PC prices set to rise as AI boom drains memory chips**

- Due to a global shortage of memory chips (driven by AI data centres), prices of smartphones and PC are set to rise; The global race to build massive AI infrastructures is squeezing supplied of components needed for mobile devices and household electronics
- The Magnificiant Seven stocks have been pledging to spend billons on construcing data centres that can power advanced AI models and products; This has prompted memory suppliers to allocate more to manuacturing resources for advanced high-bandwidth memory
- The increased operating costs are expected to be passed on to consumers; The sharpe increase in memory costs in Q4 of 2025 means that most tech hhardware companies will increase production prices, prompting declining sales in Androids and Windows PCs
- US consumers are already coming out of a bruising year marked by the impact of tariffs, inflation and unemployment; the Consumer Technology Association said in a report that consumer confidence in late 2025 hit its lowest level since July 2022




### Date:       08-Jan-2026

**Chinese Economy:  "Beijing urged to let renminbi strengthen"**

*William Sandlund in Hong Kong*

- Investors feel that China should allow the renminbi to appreciate; The currency is significantly undervalued and is fueling trade surplus; Whilst the People's Bank of China has allowed a gradual appreciation of the renminbi against the dollar, and is now trading at its strongest level since 2023, recent dollar weakness hase made CNY significantly more competitive versus other major currencies
- Interestingly, post-Trump's tarrifs on the nation CNY strengthened 4.4 per cent against the dollar, however their real inflation-adjusted rate against the currencies of its trading partners showed that a substantial depreciation since 2021 continued for much of 2025
- In real-terms, CNY is down in value 15-20 per cent since 2021 encouraging their explosive trade surplus; PBoC seems determined to ensure that CNY trades tightly within the bounds of its daily "fix" against the dollar
- Hedge fund investers believe that CNY could rise substantially against the dollar in 2026, which would be supportive of Chinese assets and mitigate critisism; The impact is felt in european exporters who since recent rises on EUR/USD and EUR/CNY have experience less competitive rates in exports
- Given that the mainland is experiencign very low inflation, if not deflation, whilst the fundamentals would would point to faster appreciation in CNY versus other majors, this would not aid in tackling disinflation - when a country is experiencing a deflationary slump their relative interest rates fo down and exchange rates depreciate


### Date:       06-Jan-2026

**Oil: "Oil prices move higher after Donald Trump’s Venezuela strikes"**

*Malcolm Moore in London and William Sandlund in Hong Kong*

- Oil prices have moved higher as a result of the US actions in Venezuela, which have created uncertainty due to their impact on global crude reserves; Brent crude traded up 1.4 per cent on monday; West Texas Intermediate was up 1.5 per cent
- The US invasion of Venezuela is purely potential based, as although the country produces less that 1 per cent of global oil outputs, Venezuela holds ~17 per cent of the world's proven crude reserves, creating opportunity to hike supply- The US President's announcement that large US oil companies would invest in Venezuela cased shares in Chevron, ConocoPhillips, and ExxonMobil to rally 4.9, 2.9, 2.1 per cent respectively
- Named 'the glut of crude' by analysts, there is scepticism regarding long-term impacts of the forced increase in crude production and supply; Investors may assume that Trump has created a hedged trade for US oil companies - increased crude via Venezuela increases general oil supply forcing global prices down, however given his backing of US oil companies in Venezuela, increased supply will have limited negative impact on US Big Oil balance sheets - this would imply that the weakest link is the European Supermajors such as Shell, TotalEnergies, BP, Eni

**Eurozone:  "Bulgaria joins Eurozone despite pro-Russian disinformation"**

*Marton Dunai in Budapest*
    
-   Bulgaria became the 21st country to adopt the euro as a primary trading currency, following Croatia who adopted it in 2023; Bulgaria has undergone recent political instability over the last five years, including but not limited to, a cabinet collapse, corrumption allegations, and the absence of a 2026 budget
-   When assessing civilian sentiment, 40 per cent support the choice to join the eurozone; the lack of support is driven by fears that local exchanges will round up prices during conversions; Far-right parties have called for a traditional referendum relating to the euro, interestingly, all countries to have ever adopted a euro referendum have denied it, including Denmark and sweden
