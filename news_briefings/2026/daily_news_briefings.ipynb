##  Article Summaries

### Date:       22-Jan-2026

**European governments turn to short-term debt as borrowing costs rise**

*Ian Smith, Senior Markets Correspondent*

- An attempts to limit the impacts of increased borrowing cost, European gorvernments are suppressing long-term sovreign bond issuance; The average expiration date of bonds issued across the eurozone is expected to drop below 10y signalling this commitment to short-dated contracts
- This month, the Netherlands completed an overhaul of their EUR 1.8bn pension system that is expected to cut demand for long-term debt; This shift has also been experienced in the UK: The reduction in defined benefit (DB) pension systems reduce demand for long-dated bonds
- Whilst short-term borrowing is typically cheaper for governments, an over reliance on short-term expiries leave public finance exposed to swings in interest rates as the debt needs to be regularly refinanced
- Japan’s ultra-long-term borrowing costs have risen above 4 per cent for the first time, as traders sell the country’s debt ahead of a snap election; A sell-off this week in Japanese debt has spread to bond markets around the world, as investors worry over record borrowing by rich nations
- The Eurozone market has also been transformed by Germany’s announcement that it would loosen its constitutional borrowing limit to make way for a huge increase in infrastructure and defence spending


### Date:       21-Jan-2026

**US green energy growth is proving hard to kill**

*Simon Mundy*

- Despite the Trump administrations vocal support for fossil fuels, the EIA predicts a 46% surge in solar generation and a 12% rise in wind over the next two years, while coal and gas-fired output are set to slide
- US electricity demand is entering its strongest growth period this century, fueled by an AI data center boom, requiring rapid power deployment that only renewables can currently provide
- Technological advances have made renewables significantly cheaper; solar costs roughly $50/MWh compared to $115 for natural gas and $194 for coal, making green energy the rational economic choice for utilities
- The intermittency argument is fading as batter storage capacity doubles; the US hit 45GW of utility-scale storage in 2025, allowing solar power to meet demand even during night cycles
- While federal policy has become "anti-green" court actions and industry coalitions (including some oil firms) are successfully challenging government efforts to block renewable projects

### Date:       20-Jan-2026

**Indonesia’s rupiah falls as Prabowo nominates nephew to central bank role**

*A. Anantha Lakshmi and Diana Mariska in Jakarta*

- The IDR weakened to Rp16,997 per USD, near 1998 crisis levels, after President Prabowo nominated his nephew (Thomas Djiwandono) as deputy governor of Bank Indonesia, reviving concerns about central bank independence
- Investors interpret the nomination as an attempt to align monetary policy with an aggressive 8% GDP growth target, risking erosion of Bank Indonesia’s autonomy and prioritisation of pro-growth finance
- Indonesia’s economy is struggling with commodity dependence, deteriorating manufacturing capacity, and limited high-income job creation, making the growth target structurally unrealistic without policy intervention
- Parliament is exploring expanding BI’s mandate beyond price stability to explicitly support growth, and granting parliament power to remove central bank officials, reinforcing concerns over institutional capture
- Prabowo’s spending plans (e.g., $28bn free school meals) have pushed the fiscal deficit to 2.92% of GDP, near the 3% legal ceiling, elevating fears of looser fiscal stance + monetary accommodation → classic fiscal dominance setup

### Date:       19-Jan-2026

**EU readies €93bn tariffs in retaliation for Trump’s Greenland threat**

*Henry Foy in Brussels and Mercedes Ruehl in Zurich*

- EU officials are preparing a €93bn tariff package targeting the US in response to Donald Trump’s threat to impose 10% levies on nations opposing his campaign to take control of Greenland
- The dispute has triggered the most severe rupture in the Western military alliance in decades, with European leaders warning that the tension poses an "existential threat" to regional security
- France and other member states are pushing to activate the EU’s "Anti-Coercion Instrument" (ACI) for the first time, a tool that could restrict US Big Tech companies and other services from accessing the European internal market
- European leaders, including Ursula von der Leyen, intend to use these retaliation threats as "carrot and stick" leverage during high-stakes private meetings with Trump at the World Economic Forum in Davos this week
- Following Trump's vow to implement tariffs by February 1, global markets reacted sharply, with Euro Stoxx 50 futures dropping 1.5% and gold surging to a record high of $4,690 per ounce


### Date:       17-Jan-2026

**Canada and China slash tariffs as they mend ties after years of dispute**

*Ilya Gridneff in Toronto and Joe Leahy in Beijing*

- The heads of state of China and Canada agreed to lift trade barriers and tarrifs on electric vehicles and canola, marking a pivotal move for international trade
- Canada agreed to accept imports of 49,000 Chinese electric cars with a 6.1 per cent tariff, down from 100 per cent
- Carney, Canada;s prime minister, aims to double exports to non-US partners over the next decade; In exchange China is slashing itslevies on Candian canola seed imports to 15 per cent, down from ~85 per cent; China will also permit visa-free travel for Canadians
- This budding relationship creats opportunities in trade, energy and agrifood for both countries; Carney said Canada was “forging new partnerships around the world to transform our economy from one that has been reliant on a single trade partner to one that is stronger and more resilient to global shocks”.

**Global Economy:   Yen tumbles and Japan stocks soar as election talk reignites ‘Takaichi trade’**

*Leo Lewis in Tokyo, Ian Smith in London and William Sandlund in Hong Kong*

- The Japanese markets are being dominated by roumours that Prime Minister Sane Takaichi will dissolve parliment on 23-Jan-2026, to call a snap election for 08-Feb-2026; The move aims to leverage her high approval ratings and solidify her party's majority; Investors are interpreting as a green lighht for aggressive fiscal stimulus and a pivot away from rapid monetary tightening

### Date:       13-Jan-2026

**UK Retail:    UK retail spending rose at lowest pace for 7 months in December**

*Valentina Romei and Philip Stafford in London*

- UK retail spending continued to growth slowly in December, below the rate of inflation, showing that the christmas season still couldn't jumpstart consumer spending; The value of retail sales increased by an annual rate of 1.2 per cent, down from 1.4 per cent in November (a new low post May's 1 per cent)
- Consumer spending in the festive season was mainly spurred by food (a Christmas un-negotiables), with spending on gifting items such as electronics, furniture, and clothing, significantly underperformed
- Retail bosses have blamed weeks of speculation over Reeves' tax rises; The value of sales rose by 2.3 per cent across 2025, according to BRC data, which is below the latest inflation rate of 3.4 percent, showing that consumers generally cut spending quantities
- A sharp slowdown in inflation in the first half of 2026, along with further rate cuts by the BoW should provide consumers a little bit of stability, and unlock real spending power

### Date:       10-Jan-2026

**UK government spending:   Spending on consultants cut by 14% across UK government**

- UK Government cut consulting costs by 14 per cent in the year ending March 2025, showing a significant attempt to reduce government spending; Home Office, in particular, reduced annual spending by 54 per cent
- A lack of government skills investment encouraged the reliance on external advisors over recent years, with civil servant numbers hitting a 20-year high; tministers have also promised to create "a leaner, higher skilled" workforce, allocating GBP 50 mn in civil service transformation funds, cutting the use of external adisors, whilst not encouraging civil servant layoffs (which would hike the unemployment rate significantly); however, they do aim to cut 1 in 10 roles by 20230
- The Department for Energy Security and Net Zero recorded the largest real-terms rise in consultancy spending of £33mn, alongside a £11mn rise in temporary worker costs; The health department accounted for the majority of the decline, delivering a GBP 1.1 bn reduction in 2024-25
- Cuts to Home Office spending follow a 10-fold increase in consultancy costs during the last parliament, driven by efforts to tackle small boat crossings and implement the Rwanda plan; This government is relentlessly rooting out waste to protect taxpayer money and make the state more efficient; This should be reflected in long-dated government bond prices

### Date:       08-Jan-2026

**Chinese Economy:  "Beijing urged to let renminbi strengthen"**

*William Sandlund in Hong Kong*

- Investors feel that China should allow the renminbi to appreciate; The currency is significantly undervalued and is fueling trade surplus; Whilst the People's Bank of China has allowed a gradual appreciation of the renminbi against the dollar, and is now trading at its strongest level since 2023, recent dollar weakness hase made CNY significantly more competitive versus other major currencies

### Date:       06-Jan-2026

**Oil: "Oil prices move higher after Donald Trump’s Venezuela strikes"**

*Malcolm Moore in London and William Sandlund in Hong Kong*

- Oil prices have moved higher as a result of the US actions in Venezuela, which have created uncertainty due to their impact on global crude reserves; Brent crude traded up 1.4 per cent on monday; West Texas Intermediate was up 1.5 per cent
- The US invasion of Venezuela is purely potential based, as although the country produces less that 1 per cent of global oil outputs, Venezuela holds ~17 per cent of the world's proven crude reserves, creating opportunity to hike supply- The US President's announcement that large US oil companies would invest in Venezuela cased shares in Chevron, ConocoPhillips, and ExxonMobil to rally 4.9, 2.9, 2.1 per cent respectively
- Named 'the glut of crude' by analysts, there is scepticism regarding long-term impacts of the forced increase in crude production and supply; Investors may assume that Trump has created a hedged trade for US oil companies - increased crude via Venezuela increases general oil supply forcing global prices down, however given his backing of US oil companies in Venezuela, increased supply will have limited negative impact on US Big Oil balance sheets - this would imply that the weakest link is the European Supermajors such as Shell, TotalEnergies, BP, Eni
