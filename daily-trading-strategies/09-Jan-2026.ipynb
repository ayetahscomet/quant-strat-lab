# Python Implementation

```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# --- Parameters ---
ticker = "LYV"
entry_price = 147.21
target_price = 174.00
stop_loss_price = 135.00
holding_period_days = 180

end_date = datetime.now()
start_date = end_date - timedelta(days=365)

print(f"Downloading data for {ticker} from {start_date.strftime('%Y-%m-%d')} to today...")

try:
    data = yf.download(ticker, start=start_date, end=end_date)

    if data.empty:
        raise ValueError("No data returned from yfinance.")

    prices = data['Close']
    if isinstance(prices, pd.DataFrame):
        prices = prices.iloc[:, 0]

    # Daily returns calculation
    returns = prices.pct_change().dropna()
    print("Data downloaded successfully.")

    latest_market_price = float(prices.iloc[-1])
    print(f"Latest {ticker} Market Price: {latest_market_price:.2f}")

    # Simulation logic for the 180-day window
    trade_pnl = []
    trade_dates = []

    i = 0
    dates = returns.index
    while i < len(returns) - holding_period_days:
        cum_return = 0
        stopped = False
        trade_start_date = dates[i]
        
        for j in range(holding_period_days):
            if i + j >= len(returns):
                break
            
            # Simple daily return compounding
            daily_ret = returns.iloc[i + j]
            cum_return = (1 + cum_return) * (1 + daily_ret) - 1

            # Adjusted stop loss based on percentage from entry
            current_stop_threshold = (stop_loss_price - entry_price) / entry_price
            if cum_return <= current_stop_threshold:
                trade_pnl.append(cum_return)
                trade_dates.append(trade_start_date)
                stopped = True
                break

        if not stopped:
            trade_pnl.append(cum_return)
            trade_dates.append(trade_start_date)
        
        i += 20 # Step forward 20 days for next simulated entry

    results = pd.DataFrame({"Start Date": trade_dates, "PnL": trade_pnl})

    if not results.empty:
        print("\n--- Strategy Summary (Backtest Context) ---")
        print(f"Average Return: {results['PnL'].mean():.2%}")
        print(f"Win Rate: {(results['PnL'] > 0).mean():.2%}")
    else:
        print("\nInsufficient historical window for 180-day holding simulation.")

except Exception as e:
    print(f"Execution Error: {e}")
```
