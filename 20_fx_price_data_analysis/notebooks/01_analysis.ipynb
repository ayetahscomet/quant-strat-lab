```python
# --- notebooks/01_analysis.ipynb ---

import pandas as pd
import matplotlib.pyplot as plt

from src.data.download_fx import download_fx, compute_returns
from src.analysis.summary_stats import compute_summary


price_df = download_fx()
returns_df = compute_returns(price_df)

summary = compute_summary(price_df, returns_df)
summary

fig, ax = plt.subplots(figsize=(12, 5))
price_df.plot(ax=ax)
ax.set_title("FX Price Series (BRL, EGP, TRY)")
plt.show()

fig, ax = plt.subplots(figsize=(12, 5))
returns_df.plot(ax=ax)
ax.set_title("Log Returns")
plt.show()

rolling_vol = returns_df.rolling(30).std()

fig, ax = plt.subplots(figsize=(12, 5))
rolling_vol.plot(ax=ax)
ax.set_title("30-Day Rolling Volatility")
plt.show()
```
