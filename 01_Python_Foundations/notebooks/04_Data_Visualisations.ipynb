```python
#   04.00-Introduction-To-Matplotlib.ipynb

import matplotlib as mpl
import matplotlib.pyplot as plt

plt.style.use('classic')

x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()

%matplotlib
%matplotlib inline

x = np.linspace(0, 10, 100)
fig = plt.figure();
plt.plot(x, np.sin(x), '-');
plt.plot(x, np.cos(x), '--');
fig.savefig('my_figure.png');

from IPython.display import Image
Image('my_figure.png');

fig.canvas.get_supported_filetypes()

# MATLAB-style Interface

plt.figure();
plt.subplot(3, 1, 1);
plt.plot(x, np.sin(x));
plt.subplot(3, 1, 2);
plt.plot(x, np.cos(x));
plt.subplot(3, 1, 3);
plt.plot(x, np.tan(x));

# Object-oriented Interface

fig, ax = plt.subplots(3);
ax[0].plot(x, np.sin(x));
ax[1].plot(x, np.cos(x));
ax[2].plot(x, np.tan(x));
```

```python
#   04.01-Simple-Line-Plots.ipynb

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x));

plt.plot(x, np.sin(x));

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));

[plt.plot(x, np.sin(x - i)) for i in range(0, 6)];
plt.show();

_linestyle = ['-', '--', '-', ':'];
for i in range(4, 8):
    for l in _linestyle:
        plt.plot(x, x + i, linestyle = l)
plt.show();

_spec = ['-g', '--c', '-.k', ':r'];
for i in range (0, 4):
    for s in _spec:
        plt.plot(x, x + i, s)
plt.show();

#   Adjusting the Plot: Axes Limits
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x));
plt.xlim(-1, 11);
plt.ylim(-1.5, 1.5);
plt.xlim(10, 0);
plt.ylim(1.2, -1.2);
plt.axis([-1, 11, -1.5, 1.5]);
plt.axis('tight');
plt.axis('equal');

plt.Figure(figsize = (10, 5));
plt.plot(x, np.sin(x));
plt.title("A Sine Curve");
plt.xlabel("x");
plt.ylabel("sin(x)");

plt.plot(x, np.sin(x), '-g', label = 'sin(x)');
plt.plot(x, np.cos(x), ':b', label = 'cos(x)');
plt.legend();
plt.show();

plt.Figure(figsize = (10, 5));
ax = plt.axes();
ax.plot(x, np.sin(x));
ax.set(xlim = (0, 10), ylim = (-2, 2),
       xlabel = 'x', ylabel = 'sin(x)',
       title = 'A Simple Plot');
plt.show();
```

```python
# 04.02-Simple-Scatter-Plots.ipynb

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
import numpy as n

x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, y, '-o', color = 'black');
plt.show();

rng = np.random.RandomState(0);
plt.Figure(figsize= (10, 5));
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label = "marker = '{0}".format(marker))
plt.legend(numpoints = 1);
plt.xlim(0, 1.8);
plt.show();

plt.plot(x, y, '-ok');

plt.plot(x, y, '-p', color = 'gray', markersize = 15, 
         linewidth = 4, markerfacecolor = 'white', 
         markeredgecolor = 'gray', markeredgewidth = 2);
plt.ylim(-1.2, 1.2);
plt.show();

plt.scatter(x, y, marker = 'o', color = 'navy');
plt.show();

rng = np.random.RandomState(0);
x = rng.randn(100);
y = rng.randn(100);
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x, y, c = colors, s = sizes, alpha = 0.3, cmap = 'viridis');
plt.colorbar();
plt.show();

from sklearn.datasets import load_iris
iris = load_iris();
features = iris.data.T;
plt.scatter(features[0], features[1], alpha = 0.2,
            s = 100 * features[3], c = iris.target, cmap = 'viridis');
plt.xlabel(iris.feature_names[0]);
plt.ylabel(iris.feature_names[1]);
plt.show();
```
