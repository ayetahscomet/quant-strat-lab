```python
def func1(a,b):
    return a / b

def func2(x):
    a = x
    b = x - 1
    return func1(a, b)

func2(5) # 1.25
func2(1) # ZeroDivisionError: division by zero
# x = 1
# b = 1- 1 = 0
# a / b = 1 / 0 = ∞ # Hence ZeroDivisionError
```
